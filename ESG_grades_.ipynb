```python
import pandas as pd
esg = pd.read_excel('/content/kcgs_esg_grades_A&B_시가총액추가.xlsx')
esg.head()
```

```python
esg.columns = ['company', 'code', 'esg', 'e', 's', 'g', 'year', 'AV']
esg.head()
```

```python
import yfinance as yf
esg.code = esg.code.map('{:06d}'.format)
esg['ret'] = 0.0
esg.head()
```

```python
for i in range(0, len(esg)):
    df = yf.download(str(esg.code[i]) + '.KS', start = '2020-01-01', end = '2020-12-31')
    ret = df.Close.pct_change().mean()
    esg['ret'][i] = ret
```

# Clustering Analysis

```python
esg_revise = esg.copy()
esg_revise.head()
```

```python
esg_revise['e']
#임시로 수치 지정
mapper = {'A+' : 12, 'A' : 10, 'B+' : 8, 'B' : 6, 'C' : 4, 'D' : 2}
esg_revise['e'] = esg_revise['e'].map(mapper)
esg_revise['s'] = esg_revise['s'].map(mapper)
esg_revise['g'] = esg_revise['g'].map(mapper)
esg_revise['esg'] = esg_revise['esg'].map(mapper)
esg_revise.head()
```

```python
import seaborn as sns
esg1 = esg_revise[['e', 's', 'g', 'AV']]
cor = esg1.corr()
sns.heatmap(cor, square = True)
```

```python
from sklearn.metrics import silhouette_score as sil
from sklearn.cluster import KMeans
all_scores = []
features = ['e', 's', 'g', 'AV']
for i in range(10):
    i = i + 2
    model = KMeans(n_clusters=i)
    model.fit(esg_revise[features])
    sil_score = sil(esg_revise[features], model.labels_)
    # 새로운 모델 만들 때마다 fit(학습) 하고 실루엣 방식으로 점수를 뽑아냄
    elbow_score = model.inertia_
    score_dict = {'cluster_num' : i, 'sil_score' : sil_score, 'elbow' : elbow_score}
    all_scores.append(score_dict)
score_df = pd.DataFrame(all_scores)
```

```python
sns.lineplot(data=score_df, x='cluster_num', y='elbow')
#기울기가 가장 급격히 감소하는 cluster num = 3
```

```python
model = KMeans(n_clusters = 3)
model.fit(esg_revise[features])
```

```python
esg_revise['cluster'] = model.labels_
esg_revise
```

```python
cluster_df = esg_revise.groupby('cluster').agg({'e' : np.mean, 's' : np.mean, 'g' : np.mean, 'AV' : np.mean})
```

```python
cluster_0 = esg_revise.loc[esg_revise['cluster'] == 0]
cluster_1 = esg_revise.loc[esg_revise['cluster'] == 1]
cluster_2 = esg_revise.loc[esg_revise['cluster'] == 2]

cluster_df['ret'] = [cluster_0['ret'].mean(), cluster_1['ret'].mean(), cluster_2['ret'].mean()]
cluster_df
```

### 지도학습(RandomForest) - feature importance 조사

```python
len(esg_revise)
```

```python
#divide continuous variable 'ret' into subintervals
esg_revise['ret_group'] = pd.qcut(esg_revise['ret'], 5, 
                                  labels = ['1', '2', '3', '4', '5'])
esg_revise
```

```python
from sklearn.ensemble import RandomForestClassifier as rf 

label = 'ret_group'
features = ['e', 's', 'g', 'AV']

model = rf()
model.fit(esg_revise[features], esg_revise[label])
```

```python
imp = pd.DataFrame()
imp['importance'] = model.feature_importances_
imp['features'] = features
imp = imp.sort_values(by = 'importance', ascending = False)
sns.barplot(data = imp, x = 'importance', y = 'features')
#시가총액의 영향이 매우 큼
```

```python
#시가총액 제외하여 model2 구축
label = 'ret_group'
features = ['e', 's', 'g']
model2 = rf()
model2.fit(esg_revise[features], esg_revise[label])
imp2 = pd.DataFrame()
imp2['importance'] = model2.feature_importances_
imp2['features'] = features
imp2 = imp2.sort_values(by = 'importance', ascending = False)
sns.barplot(data = imp2, x = 'importance', y = 'features')
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

feature_names = ['e', 's', 'g', 'AV']
target = np.array(esg_revise['ret_group'])
target = target.astype(int)
X = pd.DataFrame(esg_revise, columns = feature_names)
y = target
X.head()
```

```python
clf = DecisionTreeClassifier(max_depth=3)
clf.fit(X, y)
```
